```python
romeo_text = """ crazy flashbacks in twin peaks , even though the violence in the film pales in comparison to that in the black-and-white comic . 
oscar winner martin childs' ( shakespeare in love ) production design turns the original prague surroundings into one creepy place . 
even the acting in from hell is solid , with the dreamy depp turning in a typically strong performance and deftly handling a british accent . 
ians holm ( joe gould's secret ) and richardson ( 102 dalmatians ) log in great supporting roles , but the big surprise here is graham . 
i cringed the first time she opened her mouth , imagining her attempt at an irish accent , but it actually wasn't half bad . 
the film , however , is all good .  """
```

```python
romeo_words = nltk.word_tokenize(romeo_text)
```

```python
def build_bag_of_words_features(words):
    return {word: True for word in words}
```

```python
import string
```

```python
string.punctuation
```

```python
useless_words = nltk.corpus.stopwords.words("english") + list(string.punctuation)

```

```python
useless_words
```

```python
type(useless_words)
```

```python
def build_of_bag_of_words_features_filtered(words):
    return { 
        word: 1 for word in words \
        if not word in useless_words}

```

```python
build_of_bag_of_words_features_filtered(romeo_words)
```

```python
filtered_words= [word for word in movie_reviews.words()
                if not word in useless_words]
type(filtered_words)
```

```python
from collections import Counter
word_counter = Counter(filtered_words)
```

```python
most_common_words = word_counter.most_common()[:10]
```

```python
negative_features = [
    (build_of_bag_of_words_features_filtered
    (movie_reviews.words(fileids=[f])), 'neg')\
    for f in negative_fileids
]
```

```python
print(negative_features[3])
```

```python
positive_features = [
    (build_of_bag_of_words_features_filtered
    (movie_reviews.words(fileids=[f])), 'pos')\
    for f in positive_fileids
]
```

```python
print(positive_features[3])
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
split = 800
```

```python
sentiment_classifier = NaiveBayesClassifier.train(positive_features[:split]+negative_features[:split])
```

```python
nltk.classify.util.accuracy(sentiment_classifier, positive_features[:split]+ negative_features[:split])*100
```

```python
nltk.classify.util.accuracy(sentiment_classifier, positive_features[split:]+ negative_features[split:])*100
```

```python
sentiment_classifier.show_most_informative_features()
```
